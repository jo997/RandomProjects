```python
import pandas as pd

colnames=['input', 'empty']
df = pd.read_csv('1.csv',names=colnames, header=None)

```

```python
df
```

```python
import re

# Updated function to extract the first digit of the first number and the last digit of the last number
def extract_numbers(s):

    numbers = re.findall(r'\d+', s)

    if len(str(numbers[0])) == 1:
        # Extract the first digit of the first number
        first_number = int(numbers[0][0])
        last_number = int(numbers[-1][-1])
        return first_number, last_number
    elif numbers:
        # Extract the first digit of the first number and the last digit of the last number
        first_number = int(numbers[0][0])
        last_number = int(numbers[-1][-1])
        return first_number, last_number
    else:
        return None, None

# Apply the function
df[['first_number', 'last_number']] = df['input'].apply(extract_numbers).apply(pd.Series)
```

```python
df
```

```python
df['merged_number'] = df.apply(lambda row: int(str(row['first_number']) + str(row['last_number'])) if pd.notna(row['last_number']) else row['first_number'], axis=1)
```

```python
df
```

```python
total_sum = df['merged_number'].sum()
total_sum
```

```python
colnames=['input', 'empty']
df = pd.read_csv('1.csv',names=colnames, header=None)

word_to_digit = {
    'zero': '0',
    'one': '1',
    'two': '2',
    'three': '3',
    'four': '4',
    'five': '5',
    'six': '6',
    'seven': '7',
    'eight': '8',
    'nine': '9'
}

def replace_words_with_digits(input_string):
    for word, digit in word_to_digit.items():
        input_string = input_string.replace(word, digit)
    return input_string

# Apply the function to the 'input' column
df['input'] = df['input'].apply(replace_words_with_digits)
```

```python
df
```

```python

def extract_numbers(s):
    # Replace words with digits

    numbers = re.findall(r'\d+', s)
    if len(str(numbers[0])) == 1:
        # Extract the first digit of the first number
        first_number = int(numbers[0][0])
        last_number = int(numbers[-1][-1])
        return first_number, last_number
    elif numbers:
        # Extract the first digit of the first number and the last digit of the last number
        first_number = int(numbers[0][0])
        last_number = int(numbers[-1][-1])
        return first_number, last_number
    else:
        return None, None

# Apply the function
df[['first_number', 'last_number']] = df['input'].apply(extract_numbers).apply(pd.Series)
```

```python
import re

with open("1.csv") as f:
    data = f.read().strip()


def calibration(data):
    ls = data.split("\n")
    ns = [re.findall("\d", x) for x in ls]
    return sum(int(n[0] + n[-1]) for n in ns)


# Part 1
print(calibration(data))

# Part 2
data = (
    data.replace("one", "one1one")
    .replace("two", "two2two")
    .replace("three", "three3three")
    .replace("four", "four4four")
    .replace("five", "five5five")
    .replace("six", "six6six")
    .replace("seven", "seven7seven")
    .replace("eight", "eight8eight")
    .replace("nine", "nine9nine")
)
print(calibration(data))
```

```python
from heapq import heappush, heappop

# read in as list
data = open("input.txt", "r").read().splitlines()


# function to get neighbors of a node based on its current position, velocity, and direction
def genNeighbours(node, data, fuuu):
    x, y, v, dir = node
    dx, dy = dir
    left, right = (-dy, dx), (dy, -dx)

    # Check conditions for moving in the current direction (less tahn 3)
    if v < (10 if fuuu else 3) and 0 <= x + dx < len(data) and 0 <= y + dy < len(data[0]):
        yield (x + dx, y + dy, v + 1, dir), int(data[x + dx][y + dy])

    # Check conditions for turning left or right
    for dx, dy in left, right:
        if 0 <= x + dx < len(data) and 0 <= y + dy < len(data[0]) and (not fuuu or v > 3):
            yield (x + dx, y + dy, 1, (dx, dy)), int(data[x + dx][y + dy])


def pathfind(data, fuuu):
    # visited nodes in priority queue
    Q_visited = set()

    # Define starting nodes
    start1, start2 = (0, 0, 0, (1, 0)), (0, 0, 0, (0, 1))

    # Initialize distance dictionary for each node
    dist = {start1: 0, start2: 0}

    # Priority queue to store nodes based on their distance
    Q = [(0, start1), (0, start2)]

    # Define the target node based on the size of the input data
    target = (len(data) - 1, len(data[0]) - 1)

    # Continue exploring nodes until the priority queue is not empty
    while len(Q):
        _, u = heappop(Q)

        # Skip if the node has already been visited
        if u in Q_visited:
            continue

        # Mark the current node as visited
        Q_visited.add(u)

        # Check if the target node is reached and the condition is satisfied
        if u[:2] == target and (not fuuu or u[2] > 3):
            target = u
            break

        # Explore neighbors of the current node
        for v, cost in genNeighbours(u, data, fuuu):
            if v in Q_visited:
                continue

            # Update the distance and add the neighbor to the priority queue
            alt = dist[u] + cost
            if v not in dist or alt < dist[v]:
                dist[v] = alt
                heappush(Q, (alt, v))

    # Print the distance of the target node from the start
    print(dist[target])


pathfind(data, False)

# for part 2 set fuu to true
pathfind(data, True)
```

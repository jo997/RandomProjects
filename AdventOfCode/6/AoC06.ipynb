```python
import math
import re

# Step 1: File Reading
with open('input.txt', 'r') as f:
    input_data = f.read().splitlines()

# Step 2: Data Extraction
# Extract durations from the first line
race_durations = [int(ms) for ms in re.findall(r'\d+', input_data[0])]

# Extract distances from the second line
record_distances = [int(mm) for mm in re.findall(r'\d+', input_data[1])]

# Step 3: Brute-force Function
def number_of_ways_to_win(race_duration: int, record_distance: int) -> int:
    # Initialize the count of wins
    wins = 0
    
    # Iterate through possible holding durations
    for hold_duration in range(1, race_duration):
        # Calculate the speed and distance for the given holding duration
        speed = hold_duration
        distance = (race_duration - hold_duration) * speed
        
        # Check if the distance is greater than the record distance
        if distance > record_distance:
            wins += 1
    
    # Return the total number of wins
    return wins

# Step 4: Calculating Winning Options
# Calculate the number of ways to win for each race duration and record distance pair
winning_options_per_race = [number_of_ways_to_win(race_duration, record_distance)
                            for race_duration, record_distance in zip(race_durations, record_distances)]

# Step 5: Final Result
print(math.prod(winning_options_per_race))
```

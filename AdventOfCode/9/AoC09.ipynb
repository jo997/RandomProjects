```python
def next(seq):
    # Initialize a variable 'sum' to store the sum of the sequence
    sum = 0
    
    # Continue iterating until all elements in the sequence are zero
    while not all(a == 0 for a in seq):
        # Add the last element of the sequence to the sum
        sum += seq[-1]
        
        # Generate a new sequence by taking the differences between consecutive elements
        seq = [seq[i + 1] - seq[i] for i in range(len(seq) - 1)]
    
    # Return the final sum
    return sum

# Open the file 'input.txt' and read lines, converting each line to a list of integers
with open("input.txt") as f:
    lines = [[int(num) for num in ln.split()] for ln in f.readlines()]


# Calculate the sum of the 'next' function for each sequence in 'lines' and print the result
print(sum(next(seq) for seq in lines))

# Calculate the sum of the 'next' function for each reversed sequence in 'lines' and print the result
print(sum(next(seq[::-1]) for seq in lines))
```
